### 測試集資料，資料年份2024/01/01 ~ 2024/03/31

```python
fwd_path = os.path.abspath(os.path.join(os.getcwd(), "..")) #取得當前工作目錄的上一集目錄，並存入路徑
stock_num = ['UNIUSDT_期貨','UNIUSDT_現貨','DYDXUSDT_期貨','DYDXUSDT_現貨']
df = pd.DataFrame()
for i in stock_num:
    file_path = "C:/python/Crypto/Test_Set/"+ i +".csv"
    file = pd.read_csv(file_path, index_col = 'open_time')
    df = df.dropna()
    file = pd.DataFrame(file['close'])
    df = pd.concat([df, file], axis = 1)
df.columns = ['UNIUSDT期貨價格','UNIUSDT現貨價格','DYDXUSDT期貨價格','DYDXUSDT現貨價格']
df = df.dropna()

df['UNIUSDT_Spread'] = df['UNIUSDT期貨價格'] / df['UNIUSDT現貨價格']
df['DYDXUSDT_Spread'] = df['DYDXUSDT期貨價格'] / df['DYDXUSDT現貨價格']
df['Target_Spread'] = df['UNIUSDT_Spread'] / df['DYDXUSDT_Spread']
df['SMA'] = df['Target_Spread'].rolling(15).mean()
df['STD'] = df['Target_Spread'].rolling(15).std()
df['Upper_band'] = df['SMA'] + 2 * df['STD']
df['Lower_band'] = df['SMA'] - 2 * df['STD']
df = df.dropna()
#df.drop(df.tail(100000).index, inplace=True)
print(df)
```

```python
Spot_fund = 100000
Futures_fund = 100000                   
fee = 0.003            

position = None
profit_list = [0]                      
profit_fee_list = [0]                
Spot_executesize = 0     
Futures_executesize = 0                   

for i in range(len(df)):           
    if i == len(df)-1:             
        break

    if position == None:               
        profit_list.append(0)         
        profit_fee_list.append(0)   

        if df['Target_Spread'].iloc[i] < df['Upper_band'].iloc[i] and df['Target_Spread'].iloc[i+1] > df['Upper_band'].iloc[i+1]: 
            position = '空單'
            if (df['UNIUSDT期貨價格'].iloc[i] > df['UNIUSDT現貨價格'].iloc[i]):
                Futures_executesize = Futures_fund / (df['UNIUSDT期貨價格'].iloc[i+1] + df['DYDXUSDT期貨價格'].iloc[i+1] )
                Spot_executesize  = 5 *  Futures_executesize 

            if (df['UNIUSDT期貨價格'].iloc[i] < df['UNIUSDT現貨價格'].iloc[i]):
                Spot_executesize = Spot_fund / (df['UNIUSDT現貨價格'].iloc[i+1] + df['DYDXUSDT現貨價格'].iloc[i+1] )
                Futures_executesize  = Spot_executesize//5      

        if df['Target_Spread'].iloc[i] > df['Lower_band'].iloc[i] and df['Target_Spread'].iloc[i+1] < df['Lower_band'].iloc[i+1]:                
            position = '多單'   
            if (df['UNIUSDT期貨價格'].iloc[i] > df['UNIUSDT現貨價格'].iloc[i]):
                Futures_executesize = Futures_fund / (df['UNIUSDT期貨價格'].iloc[i+1] + df['DYDXUSDT期貨價格'].iloc[i+1] )
                Spot_executesize  = 5 *  Futures_executesize 
            else:
                Spot_executesize = Spot_fund / (df['UNIUSDT現貨價格'].iloc[i+1] + df['DYDXUSDT現貨價格'].iloc[i+1] )
                Futures_executesize  = Spot_executesize//5 

            
    elif position == '空單':                                   
        profit =  ( df['UNIUSDT期貨價格'].iloc[i] - df['UNIUSDT期貨價格'].iloc[i+1] + df['DYDXUSDT期貨價格'].iloc[i+1] - df['DYDXUSDT期貨價格'].iloc[i]) * Futures_executesize 
        + ( df['UNIUSDT現貨價格'].iloc[i+1] - df['UNIUSDT現貨價格'].iloc[i] + df['DYDXUSDT現貨價格'].iloc[i] - df['DYDXUSDT現貨價格'].iloc[i+1] ) * Spot_executesize
        profit_list.append(profit)

        if  df['Target_Spread'].iloc[i] > (0.75) * df['STD'].iloc[i] and df['Target_Spread'].iloc[i+1] < (0.75) * df['STD'].iloc[i+1] or i == len(df)-2:                    
            pl_round = ( df['UNIUSDT期貨價格'].iloc[i+1] +df['DYDXUSDT期貨價格'].iloc[i+1] ) * Futures_executesize
            + (df['UNIUSDT現貨價格'].iloc[i+1] + df['DYDXUSDT現貨價格'].iloc[i+1] ) * Spot_executesize
            profit_fee = profit - (Spot_fund + Futures_fund +pl_round) * fee
            profit_fee_list.append(profit_fee)                       
            position = None        
        else:
            profit_fee = profit
            profit_fee_list.append(profit_fee)

    elif position == '多單':                                   
        profit =  ( df['UNIUSDT期貨價格'].iloc[i+1] - df['UNIUSDT期貨價格'].iloc[i] + df['DYDXUSDT期貨價格'].iloc[i] - df['DYDXUSDT期貨價格'].iloc[i+1] ) * Futures_executesize
        + ( df['UNIUSDT現貨價格'].iloc[i] - df['UNIUSDT現貨價格'].iloc[i+1] + df['DYDXUSDT現貨價格'].iloc[i+1] - df['DYDXUSDT現貨價格'].iloc[i] ) * Spot_executesize
        profit_list.append(profit)

        if  df['Target_Spread'].iloc[i] < (-0.5) * df['STD'].iloc[i] and df['Target_Spread'].iloc[i+1] > (-0.5) * df['STD'].iloc[i+1] or i == len(df)-2:                    
            pl_round = ( df['UNIUSDT期貨價格'].iloc[i+1] +df['DYDXUSDT期貨價格'].iloc[i+1] ) * Futures_executesize
            + (df['UNIUSDT現貨價格'].iloc[i+1] + df['DYDXUSDT現貨價格'].iloc[i+1] ) * Spot_executesize
            profit_fee = profit - (Spot_fund + Futures_fund+pl_round) * fee
            profit_fee_list.append(profit_fee)                       
            position = None        
        else:
            profit_fee = profit
            profit_fee_list.append(profit_fee)            
```

```python
equity_2 = pd.DataFrame({'profit': np.cumsum(profit_list)})
print(equity_2)
equity_2.plot(figsize = (12, 6), grid = True)
A = np.cumsum(profit_fee_list)
B = np.cumsum(profit_list)
last_value2 = B[-1]
last_value = A[-1]
#print("分配金額:", fund)
print("手續費:", fee)
print("總獲利(沒扣除手續費):",last_value2)
```
